```python
# Importing libraries
import os
import re
import cv2
# import itk
import SimpleITK as sitk
import glob
import random
import shutil
import glob
import numpy as np
# import splitfolders
# import nibabel as nb
from PIL import Image
import SimpleITK as sitk
import matplotlib.pyplot as plt
import scipy.ndimage as snd
from skimage.transform import AffineTransform, warp
from skimage.filters.thresholding import threshold_otsu
```

```python
# code for image numbering
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts
```

```python
# Test Fixed Images Outer Edge Extraction
kernel = np.ones((5, 5))
input_path = r"F:\final_data\test\fixed_images_binarized" 
destination_path =r"F:\final_data\test\Eroded Edge Images"

for index, file in enumerate(sorted(glob.glob(input_path + "/*.png"), key=numericalSort)):
    image1 = cv2.imread(file)
    image2 = cv2.erode(image1,kernel)
    image3 = image1 - image2
    os.chdir(destination_path)
    plt.imsave('fixed_eroded_' + str (index) + '.png', image3, cmap='gray')
```

```python
# Test Moving Images Outer Edge Extraction
kernel = np.ones((5, 5))
input_path = r"F:\final_data\test\moving_images_binarized" 
destination_path =r"F:\final_data\test\Eroded_moving_images"

for index, file in enumerate(sorted(glob.glob(input_path + "/*.png"), key=numericalSort)):
    image1 = cv2.imread(file)
    image2 = cv2.erode(image1,kernel)
    image3 = image1 - image2
    os.chdir(destination_path)
    plt.imsave('moving_eroded_' + str (index) + '.png', image3, cmap='gray')
```

```python
# Blending the Test images 

fixed_path = r"F:\final_data\test\Eroded_fixed_images"
moving_path = r"F:\final_data\test\Eroded_moving_images"

fixed_paths = []
moving_paths = []

for file in (sorted(glob.glob(fixed_path + '/*.png'), key=numericalSort)):
    fixed_paths.append(file)

for file in (sorted(glob.glob(moving_path + '/*.png'), key=numericalSort)):
    moving_paths.append(file)


for index, (fixed, moving) in enumerate(zip(fixed_paths, moving_paths)):
    fixed_image = cv2.imread(fixed)
    moving_image = cv2.imread(moving)
    
    # Create a new image containing both the fixed and the moving images
    h, w = 240, 240
    temp = np.zeros((h, w, 3))
    temp[:,:,0] = fixed_image[:,:,0]/255.0
    temp[:,:,1] = moving_image [:,:,0]/255.0
    dest = r"F:\final_data\test\Eroded_Blended"
    os.chdir(dest)
    plt.imsave('E_blended_test_' + str(index) + '.jpg', temp) 
```
