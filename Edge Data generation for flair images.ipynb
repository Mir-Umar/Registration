```python
# code for image numbering
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts
```

```python
# Test Fixed Images Outer Edge Extraction
fixed_path = r"F:\final_data\test\fixed_images" 
moving_path = r"F:\final_data\test\Eroded_fixed_images"
destination_path =r"F:\final_data\test\Eroded_Flair_Fixed_Images"

fixed_paths = []
moving_paths = []

for file in (sorted(glob.glob(fixed_path + '/*.png'), key=numericalSort)):
    fixed_paths.append(file)

for file in (sorted(glob.glob(moving_path + '/*.png'), key=numericalSort)):
    moving_paths.append(file)


for index, (fixed, moving) in enumerate(zip(fixed_paths, moving_paths)):
    fixed_image = cv2.imread(fixed)
    moving_image = cv2.imread(moving)
    img3 = fixed_image * moving_image
    os.chdir(destination_path)
    plt.imsave('fixed_eroded_flair_' + str (index) + '.png', img3, cmap='gray')
```

```python
# Blending the Test images 

fixed_path = r"F:\final_data\test\Eroded_Flair_Fixed_Images"
moving_path = r"F:\final_data\test\Eroded_Flair_Moving_Images"

fixed_paths = []
moving_paths = []

for file in (sorted(glob.glob(fixed_path + '/*.png'), key=numericalSort)):
    fixed_paths.append(file)

for file in (sorted(glob.glob(moving_path + '/*.png'), key=numericalSort)):
    moving_paths.append(file)


for index, (fixed, moving) in enumerate(zip(fixed_paths, moving_paths)):
    fixed_image = cv2.imread(fixed)
    moving_image = cv2.imread(moving)
    
    # Create a new image containing both the fixed and the moving images
    h, w = 240, 240
    temp = np.zeros((h, w, 3))
    temp[:,:,0] = fixed_image[:,:,0]/255.0
    temp[:,:,1] = moving_image [:,:,0]/255.0
    dest = r"F:\final_data\test\Eroded_Flair_Blended"
    os.chdir(dest)
    plt.imsave('E_Fl_blnd_test_' + str(index) + '.jpg', temp)
```
